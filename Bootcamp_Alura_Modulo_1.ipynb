### Neste módulo serão utilizados dados do Sistema Único de Saúde (**SUS**) que são disponibilizados através da plataforma **DATASUS**.

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/processamento.csv', encoding='ISO-8859-1', skiprows = 3, sep=';', #encoding é usado para informar a padronização utilizada no arquivo
                    skipfooter = 12, thousands = '.', decimal = ',') # thousands e decimal são utilizados para indicar qual o separador de milhar e casas decimais usado no arquivo

dados.head()
```

```python
dados.mean()
```

# Desafio 1 - Escolher um título mais descritivo, que passe a mensagem adequada

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot.bar(x='Unidade da Federação', y='2008/Ago', figsize=(10,8)) # o padrão é utilizar ax (axis) como o nome da figura do gráfico.
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}')) # vírgula é o separador de milhares no padrão americano
plt.title('Gastos, em reais, das unidades da federação com o SUS em Agosto de 2008')

plt.show() # não mostra o resultado da função na primeira linha
```

## Com a nova base de dados referentes aos atendimentos realizados no SUS.

```python
dados_atd = pd.read_csv('/content/atendimento.csv', encoding='ISO-8859-1', skiprows = 3, sep=';', #encoding é usado para informar a padronização utilizada no arquivo
                    skipfooter = 12, thousands = '.', decimal = ',') # thousands e decimal são utilizados para indicar qual o separador de milhar e casas decimais usado no arquivo

dados_atd.head()
```

```python
dados_atd[['2008/Ago', '2008/Set']].head()
```

```python
dados_atd.mean() # somente é calculada a média das colunas em que todos os campos estão preenchidos.
```

```python
colunas_usaveis = dados_atd.mean().index.tolist() # o index aqui traz a informação a esquerda do mean, ou seja quais colunas tem todos os dados
                                                  # com o tolist transformamos a informação em uma lista do Python para conseguir adicionar uma nova coluna
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados_atd[colunas_usaveis] # atribuindo a uma variável para que possa ser trabalhado posteriormente
usaveis.head()
```

```python
usaveis = usaveis.set_index('Unidade da Federação') # definindo a colunca Unidade da Federação como o index desse slice
usaveis
```

# Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
usaveis.T.plot(figsize = (8,6))
plt.legend(bbox_to_anchor=(1.05, 1), title=('Unidades da Federação')) # bbox_to_anchor é utilizado para posicionar a legenda.
```

# Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
estados_desafio_2 = usaveis.loc[['35 São Paulo', '33 Rio de Janeiro', '41 Paraná', '26 Pernambuco', '29 Bahia']]
estados_desafio_2 = estados_desafio_2.drop(['2021/Mar', '2021/Fev'], axis=1) # droppando os meses mais recentes pois não possuem os gastos completos, distorcendo o gráfico
estados_desafio_2.T.plot(figsize = (8,6), title=('Gastos com o SUS nos estados de SP, RJ, PR, PE e BA'))
plt.show()
```
