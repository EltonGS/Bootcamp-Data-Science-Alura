# Desafio 1 - Escolher um título mais descritivo, que passe a mensagem adequada

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot.bar(x='Unidade da Federação', y='2008/Ago', figsize=(10,8)) # o padrão é utilizar ax (axis) como o nome da figura do gráfico.
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}')) # vírgula é o separador de milhares no padrão americano
plt.title('Gastos, em reais, das unidades da federação com o SUS em Agosto de 2008')

plt.show() # não mostra o resultado da função na primeira linha
```
